    <h1 style="font-size: 18px;">Reading and Primary Description of the Data 
</div>

**Explanation of Column Names:**

- **plant type**: The plant used is Vitis vinifera L. cv Pinot noir, propagated through wooden cuttings, and treated with  special preparations. S221 is a beneficial probiotic, while MR1, 15K, and A6 are used to simulate possible adverse natural impacts. The symbol ':' indicates simultaneous treatment with both the beneficial S221 and an adverse agent for a duration of 1 hour.

- **plant_id**: Identifies a certain plant from the plant_type.

- **other columns**: Morphology features per one plant:
    - **buds_total**: Amount of total buds
    - **buds_opened**: Amount of opened buds
    - **shoots_amount**: Amount of green shoots
    - **shoots_length**: Length of green shoots in cm
    - **leaves_amount**: Amount of leaves
    - **leaf_length**: Length of the longest part of a leaf in cm
    - **leaf_width**: Width of the widest part of a leaf in cm
    - **leaf_diam**: Diameter of a leaf in cm

<div style="background-color: #E8F4E3; padding: 10px;">
    <h1 style="font-size: 18px;">Calculating Additional Features from the Initial Data
</div>

### Percentage of Opened Buds Calculated from Total Buds Amount

```python
# Calculate the percentage of opened buds
grapes['buds_opened_percentage'] = (
    grapes['buds_opened']/grapes['buds_total'] * 100).apply(lambda x: round(x, 2)
)

grapes.head(3)
```

<div style="background-color: #E8F4E3; padding: 10px;">
    <h1 style="font-size: 18px;">Leaf Areas Calculated from Leaf Diameters
</div>

```python
# Calculate leaf areas by the given formula
grapes['leaf_area'] = (0.74*3.14*(grapes['leaf_diam']/2)**2).apply(lambda x: round(x, 2))
grapes.head(3)
```

<div style="background-color: #E8F4E3; padding: 10px;">
    <h1 style="font-size: 18px;">Mean Values of Features Grouped by plant_id   
</div>

```python
# Group the df by 'plant_id' and calculate means for each group
grouped = grapes.groupby('plant_id').mean()

# Round the result to 2 digits after the decimal point
grouped_rounded = grouped.round(2)

# Calculate means for the specified columns
mean_columns = [
    'shoots_length', 
    'leaves_amount', 
    'leaf_length', 
    'leaf_width', 
    'leaf_diam', 
    'leaf_area'
]

# Create new columns in the original df to store means
for col in mean_columns:
    grapes[col + '_mean'] = grapes['plant_id'].map(grouped_rounded[col])

grapes.head(5)
```

```python
# Drop the columns that are not needed for further analysis
grapes.drop(
    ['shoots_length', 
     'leaves_amount', 
     'leaf_length', 
     'leaf_width', 
     'leaf_diam'], 
    axis=1, 
    inplace=True
)

grapes.head(5)
```

```python
# Select the columns that will not be checked
columns_to_leave = [
    'buds_opened',
    'shoots_amount', 
    'buds_opened_percentage',
    'leaf_area',
    'shoots_length_mean',
    'leaves_amount_mean', 
    'leaf_length_mean',
    'leaf_width_mean', 
    'leaf_diam_mean',
    'leaf_area_mean'
]

# Drop rows with NaN values in columns other than the specified ones
grapes_short = grapes.dropna(subset=grapes.columns.difference(columns_to_leave), how='any')

grapes_short.head(4)
```

<div style="background-color:  #b9deab; padding: 10px;">
    <h1 style="font-size: 18px;">ANALYSIS OF THE DIFFERENT EXPERIMENTAL GROUPS
</div>

<div style="background-color: #eff7e6; padding: 10px;">
    <h1 style="font-size: 22px;"><u>EXPERIMENT WITH 15K AND S221</u>
</div>

**Explanation of the Experiment:**

- **aim: to determine whether the treatment with the probiotic S221 preparation can counteract any possible negative effects caused by 15K**


- **necessary conditions**: 
  - **all treatment variants or at least 15K should exhibit statistically significant differences when compared to the H2O control**
  - **15K and 15K:S221 must show significant differences between each other to validate the positive effect of S221**


- **controls**:
    - **H2O (main)**
    - **MRS (additional)**
    - **LB (additional)**
    - **LB+MRS (additional)**

**Steps of the Investigation**:
- **conduct t-tests to compare all variants with the main control, H2O;**
- **examine the impact of 15K on the morphological characteristics of grapes;**
- **explore whether the addition of S221 can neutralize the adverse effects of 15K;**
- **if a hypothesis of neutralization can be formulated, perform the t-test between 15K and the hypothetically succesful variant with 15K and S221;**
- **If the t-test between 15K and its variant with S221 yield p-values < 0.05, conclude a positive neutralizing effect of S221.**

<div style="background-color: #b9deab; padding: 10px;">
    <h1 style="font-size: 18px;">PERCENTAGE OF OPENED BUDS: 15K and S221
</div>

<div style="background-color: #E8F4E3; padding: 10px;">
    <h1 style="font-size: 18px;">Visualization and T-Test to Compare Different Variants of Grape Treatment with the Control (H2O)
</div>

```python
# List of the necessary plant types
desired_plant_types = ['15K', '15K:S221', 'S221', 'H2O', 'MRS', 'LB', 'LB+MRS']

# Filter df basing on the necessary plant types
filtered_df = grapes_research[grapes_research['plant_type'].isin(desired_plant_types)]

# Specify the column to compare
column_to_compare = 'buds_opened_percentage'

# Create a list to store the t-test results
results_buds_15K_H2O = []

# Perform t-test for each plant type
for plant_type in desired_plant_types:
    if plant_type == 'H2O':
        continue  # Skip self-comparison with 'H2O'

    h2o_values = filtered_df[filtered_df['plant_type'] == 'H2O']\
                [column_to_compare].fillna(0)
    other_plant_values = filtered_df[filtered_df['plant_type'] == plant_type]\
                        [column_to_compare].fillna(0)

    t_statistic, p_value = ttest_ind(h2o_values, other_plant_values, equal_var=False)

    significance = 'p <= 0.05' if p_value <= 0.05 else '-'

    results_buds_15K_H2O.append(
        {'Plant_Type': plant_type, 
         'T-Statistic': t_statistic, 
         'P-Value': p_value, 
         'Significance': significance}
    )

# Create a df from the results list
results_buds_15K_H2O_df = pd.DataFrame(results_buds_15K_H2O)

# Calculate mean and standard error
grouped = filtered_df.groupby('plant_type')[column_to_compare].agg(['mean', 'sem'])

# Sort the grouped df by mean in ascending order
grouped = grouped.sort_values(by='mean')

# Calculate confidence interval for each plant type
confidence = 0.95
h = grouped['sem'] * t.ppf((1 + confidence) / 2, len(grouped) - 1)
grouped['confidence_interval'] = h

# Create a bar plot with confidence intervals
plt.figure(figsize=(12, 6))

colors = ['forestgreen' if plant == 'H2O' else 'darkseagreen' for plant in grouped.index]

bar_plot = plt.bar(
         grouped.index, 
         grouped['mean'], 
         yerr=grouped['confidence_interval'], 
         capsize=6, 
         color=colors, 
         error_kw={'ecolor': 'slategray'}
)

# Set the color of the axes and frame
ax = plt.gca()  # Get the current axes
ax.spines['left'].set_color('slategray')   
ax.spines['bottom'].set_color('slategray')  
ax.spines['right'].set_color('none')       # Make right spine invisible
ax.spines['top'].set_color('none')         # Make top spine invisible

plt.title(
    'Mean Percentage of Opened Buds: 15K and S221', 
    color='slategray',
    fontsize=16
)

plt.xlabel('Plant Treatment', fontsize=14, color='slategray')
plt.ylabel('Percentage, %', fontsize=14, color='slategray')
plt.xticks(rotation=45, fontsize=14, color='slategray')
plt.yticks(np.arange(0, 101, 10), color='slategray')  # Adjust y-axis ticks 
plt.ylim(0, 100)  # Set y-axis limits 
plt.tight_layout()

# Add percentage values and confidence intervals above the bars
for idx, (mean, ci) in enumerate(zip(grouped['mean'], grouped['confidence_interval'])):
    if grouped.index[idx] == 'H2O':
        plt.text(
            idx, 
            mean + ci + 3, 
            f'control\n{mean:.2f}%\nCI: ±{ci:.2f}', 
            ha='center', 
            color='slategray', 
            fontsize=12
        )
    else:
        plt.text(
            idx, 
            mean + ci + 1, 
            f'{mean:.2f}%\nCI: ±{ci:.2f}\n{significance}', 
            ha='center', 
            color='slategray', 
            fontsize=12
        )

# Show the table of results
results_buds_15K_H2O_df
```

#### PRIMARY CONCLUSION on opened buds (15K and S221):
- **No statistically significant difference between the main control H2O and all other variants was observed.**

#### HYPOTHESIS:
- **Decrease in amount of opened buds caused by 15K could possibly be neutralized by the positive effect of S221 - variant 15K:S221. To check this hypothesis, a t-test between 15K and 15K:S221 will be conducted to ensure that there is a significant difference between 15K and 15K:S221.**

<div style="background-color: #E8F4E3; padding: 10px;">
    <h1 style="font-size: 18px;">T-Test to Compare 15K and 15K:S221 Effects
</div>

```python
# List of the necessary plant types
desired_plant_types = ['15K', '15K:S221']

# Filter df basing on the necessary plant types
filtered_df = grapes_research[grapes_research['plant_type'].isin(desired_plant_types)]

# Specify the columns 
columns_to_compare = [
    'shoots_amount', 
    'buds_opened_percentage', 
    'shoots_length_mean', 
    'leaves_amount_mean', 
    'leaf_area_mean'
]

# Create a list to store the t-test results
results = []

# Perform independent t-test for each column
for column in columns_to_compare:
    control_values = filtered_df[filtered_df['plant_type'] == '15K']\
                     [column].fillna(0)
    other_plant_values = filtered_df[filtered_df['plant_type'] == '15K:S221']\
                     [column].fillna(0)

    t_statistic, p_value = ttest_ind(control_values, other_plant_values, equal_var=False)
    
    significance = 'Significant' if p_value < 0.05 else '-'

    results.append(
        {'Column': column, 
         'T-Statistic': t_statistic, 
         'P-Value': p_value, 
         'Significance': significance}
    )

# Create a DataFrame from the results list
results_df_15K_pair_comparison = pd.DataFrame(results)
results_df_15K_pair_comparison
```

**Explanation of the Experiment:**

- **aim: to determine whether the treatment with the probiotic S221 preparation can counteract any possible negative effects caused by MR1**


- **necessary conditions**: 
  - **all treatment variants or at least MR1 should exhibit statistically significant differences when compared to the H2O control**
  - **MR1 and MR1:S221 must show significant differences between each other to validate the positive effect of S221**


- **controls**:
    - **H2O (main)**
    - **MRS (additional)**
    - **LB (additional)**
    - **LB+MRS (additional)**

**Steps of the Investigation**:
- **conduct t-tests to compare all variants with the main control, H2O;**
- **examine the impact of MR1 on the morphological characteristics of grapes;**
- **explore whether the addition of S221 can neutralize the adverse effects of MR1;**
- **if a hypothesis of neutralization can be formulated, perform the t-test between MR1 and the hypothetically succesful variant with MR1 and S221;**
- **If the t-test between MR1 and its variant with S221 yield p-values < 0.05, conclude a positive neutralizing effect of S221.**
